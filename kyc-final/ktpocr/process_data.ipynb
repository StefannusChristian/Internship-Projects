```python
import pandas as pd
```

```python
def replace_words(df, column):
    df[column] = df[column].replace({
        'ACEH (NAD)': 'ACEH',
        'NUSA TENGGARA TIMUR (NTT)': 'NUSA TENGGARA TIMUR',
        'NUSA TENGGARA BARAT (NTB)': 'NUSA TENGGARA BARAT',
        'DI YOGYAKARTA': 'DAERAH ISTIMEWA YOGYAKARTA'
    })
    return df
```

```python
data = pd.read_excel("./data_kode_wilayah.xlsx")
data = data.loc[:, ~data.columns.str.contains('^Unnamed')]
data['Kecamatan'] = data['Kecamatan'].apply(lambda x: x.upper())
data['Provinsi'] = data['Provinsi'].apply(lambda x: x.upper())
data['DaerahTingkatDua'] = data['DaerahTingkatDua'].apply(lambda x: x.upper())
data['DaerahTingkatDua'] = data['DaerahTingkatDua'].apply(lambda x: " ".join(x.split()[1:]))
data = replace_words(data, 'Provinsi')
display(data)
```
